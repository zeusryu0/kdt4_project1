```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
```

```python
lst_title = []
lst_summary = []
lst_date = []

for i in range(400): # i =페이지, 최대 400페이지
    
    def page(i):
        page = (i+1)*10-9 # i=0부터 시작하므로
        return page
    
    search = '국제유가'
    
    #기간 끝: 2022/02/10
    
    start_date = '2022.01.11' #ex) 2017.01.01
    last_date = '2022.02.10'
    
    start_d = '20220111' #ex) 20170101
    last_d = '20220210'
    
    url = 'https://search.naver.com/search.naver?where=news&sm=tab_pge&query={}&sort=0&photo=0&field=0&pd=3&ds={}&de={}&cluster_rank=25&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:from{}to{},a:all&start={}'.format(search, start_date, last_date, start_d, last_d, page(i))
    
   
    res = requests.get(url)
    soup = BeautifulSoup(res.text,'html.parser')
#   path = '#main_pack > section > div > div.group_news > ul>li'
    path = 'section > div > div.group_news > ul>li'
    lis = soup.select(path)
    
    for li in lis:
        # 제목
        
        title = li.select('div > div > a.news_tit')[0].text
        lst_title.append(title)
        print(title)
        
        # 요약
        summary = li.select('div > div.news_dsc > div > a')[0].text
        lst_summary.append(summary)  
        print(summary)
        
        # 날짜
        cnt = len(li.select('div > div > div.news_info > div.info_group > span.info'))
        print(cnt)
        try:
            if cnt >=2:
                raise
            else:
                date=li.select('div > div > div.news_info > div.info_group > span')[0].text
        except:
            try:
                date = li.select('div > div > div.news_info > div.info_group > span:nth-child(3)')[0].text
            except:
                date = li.select('div > div > div.news_info > div.info_group > span:nth-child(2)')[0].text
            
        lst_date.append(date)
        
        print(date)
        print("-"* 50)
        
        df = pd.DataFrame({'날짜':lst_date,'제목': lst_title, '요약': lst_summary})
#         df.to_csv('./oilnews_data/news_{}_{}.csv'.format(start_d,last_d))
```
